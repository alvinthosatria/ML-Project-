```python
import pandas as pd
import numpy as np

mammogram_data=pd.read_csv("mammographic_masses.data.txt", na_values="?")
mammogram_data.head()
```

```python
mammogram_data.columns=["BI-RADS Assesments", "Age", "Shape","Margin","Density","Severity"]
mammogram_data.head()
```

```python
mammogram_data.describe()
```

```python
mammogram_data.loc[(mammogram_data["Age"].isnull()|
                  mammogram_data["Shape"].isnull()|
                  mammogram_data["Margin"].isnull()|
                  mammogram_data["Density"].isnull())].head(50)
```

```python
mammogram_data=mammogram_data.drop("Density",axis=1 )
mammogram_data.head(20)
```

```python
mammogram_data.dropna(inplace=True)
mammogram_data.head(20)
```

```python
feature_data=mammogram_data[["Age", "Shape", "Margin"]].to_numpy()
label_data=mammogram_data["Severity"].to_numpy()
feature_data
```

```python
#Normalize feature data
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaledfeature_data=scaler.fit_transform(feature_data)
scaledfeature_data
```

```python
#DECISION TREES
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

clf=DecisionTreeClassifier(random_state=1)
cv_score=cross_val_score(clf, scaledfeature_data, label_data, cv=10)
print(cv_score.mean())
```

```python
#RANDOM FOREST CLASSIFIER
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
np.random.seed(112)

f_train,f_test,l_train,l_test=train_test_split(scaledfeature_data, label_data, test_size=0.2, train_size=0.8, random_state=10)
clf=RandomForestClassifier()
clf.fit(f_train, l_train)
clf.score(f_test, l_test)
```

```python
#SVM 
from sklearn import svm

svc=svm.SVC(kernel="linear",C=1.0)
cv_score=cross_val_score(svc, scaledfeature_data, label_data, cv=10)
print(cv_score.mean())
```

```python
svc=svm.SVC(kernel="poly",C=1.0)
cv_score=cross_val_score(svc, scaledfeature_data, label_data, cv=10)
print(cv_score.mean())
```

```python
svc=svm.SVC(kernel="sigmoid",C=1.0)
cv_score=cross_val_score(svc, scaledfeature_data, label_data, cv=10)
print(cv_score.mean())
```

```python
svc=svm.SVC(kernel="rbf",C=1.0)
cv_score=cross_val_score(svc, scaledfeature_data, label_data, cv=10)
print(cv_score.mean())
print("THE WINNER")
```

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
cv_scores=cross_val_score(clf, scaledfeature_data, label_data)
print(cv_scores.mean())
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier(n_neighbors=10)
cv_score=cross_val_score(clf,scaledfeature_data, label_data, cv=10)
print(cv_score.mean())
```

```python
mean_scores=[]
for i in range(1,51):
    clf=KNeighborsClassifier(n_neighbors=i)
    cv_score=cross_val_score(clf,scaledfeature_data, label_data, cv=10)
    mean_scores.append(cv_score.mean())
mean_scores.sort(reverse=True)
print(mean_scores[0])
```

```python
#NAIVE BAYES
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
clf=MultinomialNB()
scaler=MinMaxScaler()

minmax_scaledfeature=scaler.fit_transform(scaledfeature_data)
cv_score=cross_val_score(clf, minmax_scaledfeature, label_data, cv=10)
print(cv_score.mean())
```
